```python
rows_before_target_removal = len(df)

df = (
    df.dropna(subset=["pm25_concentration"])
    .copy()
    .reset_index(drop=True)
)

rows_removed_missing_target = rows_before_target_removal - len(df)

feature_cols = [
    "pm10_concentration",
    "no2_concentration"
]

missing_before_imputation = df[feature_cols].isnull().sum()

# Sort chronologically within each city before interpolation.
df = (
    df.sort_values(["country_name", "city", "year"])
    .reset_index(drop=True)
)

for col in feature_cols:
    # 1. Interpolate observations from the same city over time.
    df[col] = df.groupby(
        ["country_name", "city"]
    )[col].transform(
        lambda x: x.interpolate(
            method="linear",
            limit_direction="both"
        )
    )

    # 2. Fill remaining values with the country-and-year median.
    df[col] = df[col].fillna(
        df.groupby(
            ["country_name", "year"]
        )[col].transform("median")
    )

    # 3. Fill remaining values with the country median.
    df[col] = df[col].fillna(
        df.groupby("country_name")[col].transform("median")
    )

    # 4. Use the global median only as the final fallback.
    df[col] = df[col].fillna(df[col].median())

print("Rows removed because PM2.5 was missing:", rows_removed_missing_target)

print("\nMissing predictor values before imputation:")
print(missing_before_imputation)

print("\nMissing values after final cleaning:")
print(df.isnull().sum())
```

```python
pollutant_col=["pm10_concentration","pm25_concentration","no2_concentration"]

df[pollutant_col] = df[pollutant_col].round(2)
```

```python
pollutant_cols = [
    "pm10_concentration",
    "pm25_concentration",
    "no2_concentration"
]

negative_counts = (df[pollutant_cols] < 0).sum()
duplicate_count = df.duplicated().sum()

print("\nNegative pollutant values:")
print(negative_counts)

print("\nDuplicated rows before final removal:")
print(duplicate_count)

# Remove exact duplicates, if any were created after dropping unused columns.
df = df.drop_duplicates().reset_index(drop=True)

if df[pollutant_cols].isnull().any().any():
    raise ValueError("Missing pollutant values remain after cleaning.")

if (df[pollutant_cols] < 0).any().any():
    raise ValueError("Negative pollutant values remain after cleaning.")

if df["year"].isnull().any():
    raise ValueError("Missing year values remain after cleaning.")

print("\nFinal dataset shape:")
print(df.shape)

print("\nFinal missing values:")
print(df.isnull().sum())
```
